```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
ohe=OneHotEncoder()
df=pd.DataFrame(ohe.fit_transform(df).toarray())
df.head()
```

```python
input_data=df.iloc[:,:-1]
output_data=df.iloc[:,-1]
```

```python
ss=StandardScaler()
input_data=pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)
```

```python
input_data
output_data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
import tensorflow as tf
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann=Sequential()
```

```python
ann.add(Dense(units=6,activation='relu'))
ann.add(Dense(units=4,activation='relu'))
ann.add(Dense(units=2,activation='relu'))
ann.add(Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=8,epochs=100,validation_data=(x_test,y_test))
```

```python
prd=ann.predict(x_test)
prd_df=[]
for i in prd:
    if i>0.5:
        prd_df.append(1)
    else:
        prd_df.append(0)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,prd_df)
```

```python
prd1=ann.predict(x_train)
prd1_df=[]
for i in prd1:
    if i>0.5:
        prd1_df.append(1)
    else:
        prd1_df.append(0)

```

```python
accuracy_score(y_train,prd1_df)
```
